```
# crop한 파일 불러오기, 저장하기

import os
import csv
import random

mp4list = os.listdir("/content/drive/MyDrive/slowfast/Data/Kinetics/final2_normal")
mp4list1 = os.listdir("/content/drive/MyDrive/slowfast/Data/Kinetics/final2_assult")
mp4list2 = mp4list1[:1700]
mp4list3 = os.listdir("/content/drive/MyDrive/slowfast/Data/Kinetics/robber_1700")


random.shuffle(mp4list)
random.shuffle(mp4list2)
random.shuffle(mp4list3)

print("normal 파일명:  ",mp4list)
print("normal 파일 개수:  ",len(mp4list))
print("assult 파일명:  ",mp4list2)
print("assult 파일 개수:  ",len(mp4list2))
print("robber 파일명: ",mp4list3)
print("robber 파일 개수: ",len(mp4list3))

count = 0
count2 = 0
count3 = 0
train_rate = 0.8
val_rate = 0.1
test_rate = 0.1


train_count = int(len(mp4list) * train_rate)
val_count = int(len(mp4list) * val_rate)
test_count = int(len(mp4list) * test_rate)

train_count2 = int(len(mp4list2) * train_rate)
val_count2 = int(len(mp4list2) * val_rate)
test_count2 = int(len(mp4list2) * test_rate)

train_count3 = int(len(mp4list3) * train_rate)
val_count3 = int(len(mp4list3) * val_rate)
test_count3 = int(len(mp4list3) * test_rate)

print("normal train:",train_count)
print("normal val:",val_count)
print("normal test:",test_count)

print("assult train:",train_count2)
print("assult val:",val_count2)
print("assult test:",test_count2)

print("robber train:",train_count3)
print("robber val:",val_count3)
print("robber test:",test_count3)

# normal 파일 train,val,test csv로 나누기
for name in mp4list:
    path = "/content/drive/MyDrive/slowfast/Data/Kinetics/final2_normal/" + name
    if count < train_count :
         f = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_train/train1.csv" , 'a', encoding='utf-8', newline='')
         wr= csv.writer(f)
         wr.writerow([path, 0])
    elif count >= train_count and count <= val_count+train_count:
        f = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_val/val1.csv" , 'a', encoding='utf-8', newline='')
        wr = csv.writer(f)
        wr.writerow([path, 0])
    elif count > val_count+train_count and count <= train_count+val_count+test_count :
        f = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_test/test1.csv" , 'a', encoding='utf-8', newline='')
        wr = csv.writer(f) 
        wr.writerow([path, 0])
    count += 1


    f.close()

# assult 파일 train,val,test csv로 나누기
for name in mp4list2:
    path = "/content/drive/MyDrive/slowfast/Data/Kinetics/final2_assult/" + name
    if count2 < train_count2 :
         f2 = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_train/train2.csv" , 'a', encoding='utf-8', newline='')
         wr= csv.writer(f2)
         wr.writerow([path, 1])
    elif count2 >= train_count2 and count2 <= val_count2+train_count2:
        f2 = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_val/val2.csv" , 'a', encoding='utf-8', newline='')
        wr = csv.writer(f2)
        wr.writerow([path, 1])
    elif count2 > val_count2+train_count2 and count2 <= train_count2+val_count2+test_count2 :
        f2 = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_test/test2.csv" , 'a', encoding='utf-8', newline='')
        wr = csv.writer(f2) 
        wr.writerow([path, 1])
    count2 += 1


    f2.close()

# robber 파일 train, val, test csv로 나누기
for name in mp4list3:
    path = "/content/drive/MyDrive/slowfast/Data/Kinetics/robber_1700/" + name
    if count3 < train_count3 :
         f3 = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_train/train3.csv" , 'a', encoding='utf-8', newline='')
         wr= csv.writer(f3)
         wr.writerow([path, 2])
    elif count3 >= train_count3 and count3 <= val_count3+train_count3:
        f3 = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_val/val3.csv" , 'a', encoding='utf-8', newline='')
        wr = csv.writer(f3)
        wr.writerow([path, 2])
    elif count3 > val_count3+train_count3 and count3 <= train_count3+val_count3+test_count3 :
        f3 = open("/content/drive/MyDrive/slowfast/Data/Kinetics/merge_test/test3.csv" , 'a', encoding='utf-8', newline='')
        wr = csv.writer(f3) 
        wr.writerow([path, 2])
    count3 += 1
    
    f3.close()
```

```
# train, val, test csv파일 병합하기

import pandas as pd
import glob
import os

#merge train.csv
input_path = '/content/drive/MyDrive/slowfast/Data/Kinetics/merge_train' # csv파일들이 있는 디렉토리 위치
output_file = '/content/drive/MyDrive/slowfast/Data/Kinetics/merge_train/merge_train.csv' # 저장 파일명

file_list = glob.glob(os.path.join(input_path, '*.csv')) # 모든 csv파일 선택

with open(output_file, 'w') as write_file:
	for i, file_name in enumerate (file_list):
		with open(file_name, 'r') as original_file: # 파일 읽어오기
			n = 0
			while True:
				line = original_file.readline()
				#if n == 0: # 모든 csv 파일 릴레이션 삭제
				if i != 0 and n == 0: # 첫 번쨰 파일이 아니고, 첫 번쨰 라인인 경우 (릴레이션)
					n += 1
					continue
				if not line:
					print(f'File[{i}] Succeed..!')
					break
				write_file.write(line)
			n += 1

print('File Mergin Succeed..!')

#merge val.csv
input_path = '/content/drive/MyDrive/slowfast/Data/Kinetics/merge_val' # csv파일들이 있는 디렉토리 위치
output_file = '/content/drive/MyDrive/slowfast/Data/Kinetics/merge_val/merge_val.csv' # 저장 파일명

file_list = glob.glob(os.path.join(input_path, '*.csv')) # 모든 scv파일 선택

with open(output_file, 'w') as write_file:
	for i, file_name in enumerate (file_list):
		with open(file_name, 'r') as original_file: # 파일 읽어오기
			n = 0
			while True:
				line = original_file.readline()
				#if n == 0: # 모든 csv 파일 릴레이션 삭제
				if i != 0 and n == 0: # 첫 번쨰 파일이 아니고, 첫 번쨰 라인인 경우 (릴레이션)
					n += 1
					continue
				if not line:
					print(f'File[{i}] Succeed..!')
					break
				write_file.write(line)
			n += 1

print('File Mergin Succeed..!')

# merge test.csv
input_path = '/content/drive/MyDrive/slowfast/Data/Kinetics/merge_test' # csv파일들이 있는 디렉토리 위치
output_file = '/content/drive/MyDrive/slowfast/Data/Kinetics/merge_test/merge_test.csv' # 저장 파일명

file_list = glob.glob(os.path.join(input_path, '*.csv')) # 모든 scv파일 선택

with open(output_file, 'w') as write_file:
	for i, file_name in enumerate (file_list):
		with open(file_name, 'r') as original_file: # 파일 읽어오기
			n = 0
			while True:
				line = original_file.readline()
				#if n == 0: # 모든 csv 파일 릴레이션 삭제
				if i != 0 and n == 0: # 첫 번쨰 파일이 아니고, 첫 번쨰 라인인 경우 (릴레이션)
					n += 1
					continue
				if not line:
					print(f'File[{i}] Succeed..!')
					break
				write_file.write(line)
			n += 1

print('File Mergin Succeed..!')
```
