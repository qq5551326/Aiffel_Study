# 14-3. Deep Learning == Representation Learning ? (2) Representation Learning

그렇다면 '내재적 표현', 혹은 '표현(Representation)'은 무엇을 의미할까요?

다음과 같은 물체가 있다고 해 봅시다.



우리 눈에 꽃의 이미지가 보이는군요. 보다 엄밀하게 이야기하기 위해 위에 보이는 것을 '꽃'이라는 표현 대신 '데이터'라고 지칭하겠습니다.

위의 데이터는 다양한 방법으로 '표현' 될 수 있습니다. 몇 가지 다른 방법으로 위 데이터를 표현해 보죠.

총 네 가지 방법으로 왼쪽의 데이터를 표현해 보겠습니다.

1. 데이터는 분자 형태로 표현될 수 있습니다. 아니, 사실 분자 형태는 실제 데이터(물체) 그 자체라고도 할 수 있죠. 아무런 가공을 거치지 않은, 데이터가 존재하는 형태 그대로의 표현입니다.


2. 혹은 이미지로 표현해 볼 수도 있습니다. 분자 형태, 혹은 데이터 그 자체보다는 아주 조금 가공된 형태입니다. '시각적' 표현이라고도 할 수 있겠습니다. 보통 이미지를 컴퓨터에서 표현할 때에는 (14x14)와 같이 특정 사이즈를 가지는 2차원 배열로 표현됩니다. 컬러 이미지라면 R, G, B의 3채널을 가지기 때문에 3차원 배열로 표현하기도 하죠. 디지털 세상에서 RGB 색상은 이 링크처럼 Red, Green, Blue 세 가지의 색의 조합으로 나타납니다. 색의 3원색을 각각 어느 정도로 섞느냐에 따라 무지갯빛 색상을 모두 만들어낼 수 있죠. 따라서 컬러 이미지는 각 픽셀 포인트마다 Red, Green, Blue, 총 세 가지의 값을 가지므로 (14x14) 크기의 2차원 배열을 Red 용, Green 용, Blue 용 총 세 장 사용해 (14x14x3)의 3차원 배열로 표현합니다. 물론 흑백 이미지라면 RGB가 필요없이 흑과 백의 밝은 정도를 0~255 사이의 수치로만 표현하면 되므로 (14x14) 한 장, 즉 2차원 배열로 표현됩니다.


3. 세 번째 표현 방법은 표로 나타내는 것입니다. 해당 데이터가 가지고 있는 '특징(Feature)'을 각 열에 나타내고, 해당 열마다 어떤 값을 가지는지를 표 형식으로 나타낼 수 있죠. 다만, 표 형식으로 나타낼 때에는 먼저 해당 데이터를 '어떤 특징(Feature)으로 나타낼 것인지', 즉 각 열의 내용을 적합한 것들로 정의하는 과정이 필요합니다. 위와 같은 데이터를 표로 나타내려고 할 때 '손가락의 길이'와 같은 열이 있으면 안 되겠죠. 즉, 데이터를 표로 나타내기 위해서는 사람이 개입해야 합니다.


4. 마지막 표현 방법은 카테고리입니다. 카테고리는 머신러닝에서 주로 '예측하고자 하는 값'으로 사용되는 것이기도 합니다. 데이터를 카테고리로 표현하려면 '어떤 카테고리 공간이 있는지'와 '그중 어떤 카테고리에 속하는지'를 정하는 과정이 필요합니다. 즉, 카테고리 공간이 ['무궁화', '수국', '진달래', '카네이션']이라면 그중 '무궁화'라는 알맞은 카테고리로 표현할 수 있습니다. 하지만 만약 카테고리 공간이 ['판다', '호랑이', '거북', '강아지']로 되어 있다면 그 안에서 아무리 잘 선택해봐야 '판다' 혹은 '호랑이' 정도로 밖에 표현할 수 없습니다. 즉, 카테고리를 활용한 표현은 사람의 개입이 매우 강하게 들어갑니다. 사람이 카테고리 공간을 잘 정의해놔야, 해당 데이터가 알맞은 카테고리로 표현될 수 있습니다.


네, 위와 같이 네 가지 방법으로 데이터를 표현해 보았습니다. 이게 대체 무슨 의미가 있길래 이렇게 설명하는 걸까요?

위의 네 가지 표현 방법은 언뜻 보기에는 아무런 관련이 없어 보일 수 있지만, 사실 아주 강력한 계층적(Hierarchical) 인 관계를 갖고 있습니다.
무슨 말인고 하니, 1번부터 4번으로 갈수록 사람의 개입이 강하게 들어가는 표현이라는 겁니다.

다음과 같이 나타낼 수 있죠.

위에 있는 분자 형태와 이미지의 표현은 Raw한 Sensory한 표현이라고 할 수 있습니다. 특별한 가공이 들어가지 않은 날것의, 그저 감각기관만으로 받아들일 수 있는 데이터이죠.

반면, 표와 카테고리의 아래쪽 표현들은 Abstract하고 Internal한 표현들입니다. 사람의 개입이 강하게 들어가고, 그렇기 때문에 더 추상적이고 함축적인 내재적 표현을 할 수 있게 되죠.

# 14-4. 딥러닝을 관통하는 철학 (1) 행동주의
***
**조금 더 파고들어보자, 딥러닝을 관통하는 철학 속으로**  
딥러닝의 목표는 데이터의 내재된 표현을 잘 추출하는 것이라는 점을 알았습니다.

그렇다면 이러한 문제의식은 어디서 왔을까요? 조금 더 깊은 철학으로 들어가 보겠습니다.

딥러닝의 본질에 대해 논하려면 'Connectionism(연결주의)'이라고 불리는 인지과학의 한 학파로 거슬러 올라가야 합니다. 인지과학은 인간의 지능을 자극과 반응의 관계로만 설명하던 행동주의 심리학에 반발하면서 등장한 학문이죠.

연결주의와 인지과학, 행동주의.. 이게 다 무슨 이야기일까요! 천천히 하나씩 살펴보겠습니다.


행동주의: 학습된 행동은 자극의 반복으로 만들어진다

행동주의는 인간의 지능이 '자극 → 행동' 과정의 반복으로 만들어진다고 주장합니다.


행동주의는 인간, 혹은 인간뿐만 아니라 모든 생명체의 지능은 반복되는 자극에 따라 형성된다고 주장하였습니다.  
행동주의를 연구한 대표적인 학자 스키너(Skinner)는 비둘기에게 적절한 보상을 줌으로써 탁구를 학습시키는 것이 가능하다는 것을 실험으로 보여주었죠.

이렇게 행동주의는 생명체가 학습하는 과정을 단순하게 자극과 행동의 반복으로 이루어지는 것이라고 설명합니다.

이 중 생명체가 자신에게 유리한 결과를 가져다주는 행동을 알게 된다면 그 행동의 빈도를 높인다는 '강화이론'은, 머신러닝의 방법론 중 '강화학습(Reinforcement Learning)'의 근간이 되는 이론이기도 합니다. 강화학습은 실제로 특정 행위자를 학습시키기 위해 행위자가 하는 행동이 얼마나 좋았는지에 대한 척도인 '보상'을 부여하고, 행위자는 그 보상을 높이기 위해 자신의 행동을 교정하면서 학습해나가게 되죠.

# 14-5. 딥러닝을 관통하는 철학 (2) 인지주의

**인지주의: 인간은 비둘기와는 다르다! 정보를 처리하는 '의식'이 있다**  

하지만 행동주의는 인간의 내면을 지나치게 단순화했다는 비판을 받았습니다. 단순히 반복되는 자극에 따라 행동을 '강화'함으로써 학습되는 것이 아니라, 인간의 뇌 안에서 기억, 주의 등의 일련의 정보처리 과정을 거친 후 행동을 만들어낸다는 주장이 노암 촘스키(Noam Chomsky)에 의해 새롭게 제기되었죠. 이렇게 행동주의를 반대하며 제기된 학문이 바로 인지심리학입니다.

인지주의는 자극과 반응 사이에 있는 '정보처리' 과정에 관심을 두었습니다. 생명체의 지능은 단순히 어떤 행동을 자극에 맞추어 '강화'하는 것을 넘어서서, 자극을 받아 반응을 하기까지 복잡한 알고리즘을 통해 정보를 처리하는 것이라고 생각한 거죠. 다음과 같은 그림으로 나타낼 수 있습니다.

이런 블랙박스 정보 처리는 인간의 어떤 부분에서 일어날까요?

바로 우리의 뇌입니다. 우리는 분명 눈앞에 어떤 상황이 벌어진다거나, 혹은 물리적으로 신체에 어떤 자극이 왔을 때 머릿속으로 여러 가지 생각을 합니다. 행동주의에 따르면 우리가 외부에서 자극을 받았을 때 단순히 이전에 받은 보상의 경험에 따라서만 행동하겠죠. 하지만 그런 관점으로는 지금 우리가 이 글을 읽으면서 머릿속으로 이해를 하거나, 혹은 여러 가지 관련된 상상을 하는 것에 대해 설명하지 못합니다.

여러분은 행동주의와 인지주의 중 어떤 것이 더 설득력이 있는 것 같나요?

# 14-6. 딥러닝을 관통하는 철학 (3) 연결주의

**연결주의: 정보처리는 뇌에 있는 수많은 뉴런의 연결고리를 통해 이루어진다**  

인지주의에는 위와 같이 자극으로부터 행동을 만들어내는 정보처리 과정을 설명하는 모형이 몇 가지 있습니다. 기호 모형(Symbolic modeling) 혹은 연결주의 모형(Connectionism) 등이 있죠.

이들은 페드로 도밍고스(Pedro Domingos)가 쓴 '마스터 알고리즘'이라는 책에서 소개되는 머신러닝의 다섯 종족과도 연결됩니다. 다섯 종족은 기호주의자, 연결주의자, 진화주의자, 베이즈주의자, 그리고 유추주의자입니다. 각각에 대해서는 오늘 소개하고자 하는 내용을 벗어나므로 자세히 다루지는 않겠지만, 머신러닝과 딥러닝의 철학에 대해 관심이 있으시다면 아래 책을 읽어보시는 것을 추천합니다. 수학적 접근을 배제하고, 이야기로 풀어나가는 점이 인상적인 책이죠.

사족이 길었네요. 아무튼, 인지주의에는 자극으로부터 반응을 하기까지 정보를 처리하는 과정을 수행하는 모형이 몇 가지 있습니다.
그중 우리가 주목해야 할 것은 바로 연결주의(Connectionism) 입니다. 연결주의는 몇 백억 개의 뉴런이 연결되어 있는 뇌의 형태처럼, 수많은 신호들이 연결된 일련의 과정에 따라 정보가 처리된다고 보죠.

* 연결주의는 자극으로부터 반응을 하기까지의 과정에서 뉴런과 같이 연결되어 있는 모형이 정보를 처리해 나간다고 설명한다.  


* 이에 따라 연결주의의 지능체는 처음에는 '백지' 상태이며, 다수의 사례를 주고 '경험'함으로써 스스로 천천히 '학습'해 간다.  


* 즉, 학습된 내용은 연결된 뉴런 자체에 저장되어 있으며, 외부에서 자극을 받음에 따라 그 연결 형태가 바뀌면서 학습된 내용이 바뀌어간다.

# ⅱ. 함수의 역할 두 번째, Transformation: x를 변환해 주는 도구
***


함수의 역할 두 번째는 x를 변환해 주는 도구라고 할 수 있습니다.  
함수를 변환(Transformation)의 관점에서 볼 때, 우리가 필수적으로 이해해야 하는 개념은 바로 선형 변환(Linear Transformation)입니다.  

* Linear Transformation이 행렬곱 연산과 연관되어 있다고 말한다.
* 영상에서는 행렬곱 연산을 단순히 곱셈과 덧셈으로 이루어지는 계산이 아닌, 선형 변환의 관점에서는 어떤 의미를 갖는지를 설명하고자 한다.

* Transformation은 function(함수)의 fancy한 용어라고 설명한다. (즉, 같은 용어이다.)
* 특히, 선형대수학에서는 이러한 '변환' 역할을 하는 함수가 변환하기 전의 vector를 입력받아 변환된 후의 vector를 출력한다.

2차원 평면에서의 Transformation은 축(grid)을 '옮기는(move)' 역할을 한다. 어떤 변환을 하냐에 따라 수직과 수평이었던 2차원 축을 옮길수도, 휘게 할 수도, 꼬이게 할 수도 있다.

첫 번째, 변환된 축은 여전히 '직선' 형태를 띄어야 한다. 즉, 휘거나 구부러지면 안된다.

두 번째, 변환 후에도 원점(origin)은 이동하지 않아야 한다. 즉, 원점은 그대로여야 한다.

```python
import numpy as np

def linear_transformation(X, A):
    transformed_vector = np.matmul(X, A)
    return transformed_vector
```

```python
# 코드를 작성해주세요.
X = np.array([-1, 2])

A = np.array([[1, -2],
              [3, 0]])

linear_transformation(X, A)
```

***
마지막으로 함수를 x의 공간에서 y의 공간으로 매핑(Mapping) 해주는 도구라는 관점에서 접근해 보겠습니다.

사실 바로 전에 본 변환(Transformation)의 관점 또한, 하나의 벡터를 다른 벡터로 매핑하는 것이라고 볼 수 있습니다. 다만, 이번에는 조금 더 다양한 매핑을 살펴보려고 합니다.

바로 함수의 입력과 출력이 스칼라(Scalar)인지, 벡터(Vector)인지에 따라 구분해 보는 것이죠.

스칼라와 벡터는 아마 친숙한 개념일 것 같습니다. 다음 글들을 보고 간단히 알아봅시다.

* 스칼라는 '크기'만 가지고 '방향'은 가지지 않는 양이다. 주로 단 하나의 숫자로 표현한다.

# ✓ Many-to-One  
Many-to-One은 여러 가지의 정보를 입력받아 하나의 값을 출력하는 경우입니다.

이 경우 함수는 다음과 같이 벡터인 xx를 입력받아 스칼라인 yy를 출력하는 형태로 나타낼 수 있습니다. 이때 일반적으로 벡터는 볼드체로 \mathbf xx와 같이 나타냅니다.

이제 함수가 이전보다 많은 데이터를 입력받고 있죠. 벡터를 입력받아서, 단 하나의 스칼라를 출력하는 형태입니다.

사족을 덧붙이자면, 일반적으로 머신러닝/딥러닝에서는 이러한 문제를 회귀(Regression) 문제라고 합니다. 더 정확하게 회귀 문제란 입력 데이터가 몇 개인지는 상관없이, 단 하나의 연속된 실수 스칼라값을 맞추는 문제를 일컫습니다. 그러니까 위에서 이야기했던 One-to-One 문제도 회귀 문제라고 할 수 있죠. 정확히는 One-to-One의 경우 입력되는 변수가 단 하나이기 때문에 단변량 회귀(Univariate Regression), 그리고 Many-to-One의 경우에는 입력 변수가 여러 개이므로 다변량 회귀(Multivariate Regression)라고 합니다.

여기서 함수는 성격의 여러 가지 특징을 숫자로 표현한 벡터를 입력받은 후 4개의 숫자로 이루어진 벡터를 출력함으로써 MBTI 결과를 만들어내는 것이죠. 실제로 아주 많은 사람들의 설문 결과와 성격 유형에 대한 데이터가 있다면 위와 같은 성격 유형 예측 모델을 만드는 것도 가능할 것입니다.

우리가 두 점 (1, 2)(1,2), (0, 5)(0,5) 를 지나는 이차 함수를 구하는 문제를 푼다면 f(x) = ax^2 + bx + cf(x)=ax

이 경우 저 두 점을 지나는 정확한 함수가 정해져 있기 때문에 aa와 bb, cc도 단 한 가지 값으로 정해집니다.

하지만 우리가 머신러닝, 혹은 딥러닝에서 다루어야 하는 함수는 정확히 단 하나로 정해져있는 함수도 아닐뿐더러, 그 함수가 이차 함수인지, 삼차 함수인지, 혹은 신경망 형태로 나타날지조차 알지 못합니다. 우리에게 주어진 것이라곤 단 하나! 데이터 뿐이죠.

(여기서 잠깐, 신경망이라는 함수는 일차 함수일까요, 이차 함수일까요, 아니면 어떤 형태의 함수일까요? 이 물음에 대한 답은 바로 다음에 다룹니다. 그전에 한번 생각해 보세요!)

집값을 맞추는 문제를 풀 때 우리에게는 집의 평수, 위도와 경도, 층수 등 각 집의 특징을 나타내는 데이터가 있을 겁니다. 이러한 특징 데이터를 함수(모델)에 입력했을 때 우리는 모델이 집의 가격을 정확히 맞추기를 바라죠. 하지만 현실 세계에서 집의 가격은 아주 복잡한 요인들이 관여합니다 (아마 현 시기의 경제 상황이라던가, 그런 설명하기 복잡한 것들까지 반영되었을 겁니다!). 그런 모든 요인을 정량화하여 함수에 넣어줄 수 없다면 우리는 완벽한 함수는 만들어낼 수 없을 것입니다.

그렇기에 머신러닝과 딥러닝은 '완벽한 함수'를 찾는 수학 문제를 푸는 것이 아니라, '그나마 가장 잘 근사할 수 있는 함수'에 조금씩 가까워지도록 시도하는 것에 가깝습니다.

다양한 모델(함수) 중 어떤 함수를 채택하느냐는 모델의 최종 성능에 굉장히 중요한 영향을 미칩니다.

만약 간단한 선형 함수만으로 데이터를 잘 표현할 수 있다고 판단된다면 선형함수를 사용해도 되지만, 이미지나 자연어같이 형태가 고정되지 않은 '비정형 데이터'같은 경우 최근에는 거의 필연적으로 신경망을 사용합니다. 신경망의 구조는 여러 계층이 쌓인 형태이며 이렇게 'Hierarchial(계층적인)' 구조가 비정형 데이터의 표현을 단계별로 잘 추출해낼 수 있다는 것이 실험적으로 보여졌기 때문입니다.

이렇게 어떤 데이터에 대한 문제를 풀기 위해 '이러한 형태의 함수가 유리할 것'이라고 판단하고 함수 공간을 정하는 것을 Inductive Bias, 혹은 Prior를 가정한다고 합니다.

Inductive Bias란 데이터를 설명할 수 있는 최적의 함수가 특정한 함수 공간에 존재할 것이라는 가설을 말합니다.

예를 들어, 우리가 함수 공간을 f(x) = wx+bf(x)=wx+b로 일차 함수 형태로 정해놓고 시작한다면 우리는 데이터를 더 잘 표현할 수 있는 f(x) = x^2 + 5x + 2f(x)=x 
2
 +5x+2와 같은 이차함수는 영영 못 찾게 될 겁니다. 혹은, 이미지나 자연어와 같은 데이터를 처리할 때에 신경망의 형태로 모델을 정해놓고 시작하는 것은 신경망이 비정형 데이터를 표현하는 데에 강력하다는 실험적 결과에 따른 가설에 입각하는 것이죠.

이러한 Inductive Bias가 중요한 이유는 우리가 학습시킬 모델이 세상의 모든 데이터를 볼 수 없기 때문입니다. 모델이 학습하는 과정에서 보지 못한 데이터에 대해서도 예측을 잘 하려면(즉, 일반화(Generalization)가 잘 되려면), 일반적인 패턴을 잘 반영할 수 있는 형태의 모델이어야 합니다. 그렇기에 어떤 문제를 풀고자 할 때 알맞은 함수 형태로 모델을 선택하는 것은 그 데이터에 대한 도메인 지식을 가지고 있는 사람의 몫이며, 더 좋은 모델을 찾기 위해서는 그 데이터에 알맞은 함수 공간, 즉 Inductive Bias를 잘 설정하는 것이 중요합니다.

2. 모델의 함수 형태를 정했다면, 해당 함수 공간 안에서 최적의 함수를 찾아나갑니다.
모델을 어떤 함수 형태로 사용할 것인지 정했다면, 그 후로는 그 함수 공간 안에서 찾을 수 있는 최적의 함수로 학습을 해 나갑니다.

이 단계가 보통 머신러닝/딥러닝에서 말하는 '모델 학습'에 해당하죠.
머신러닝 모델을 '학습'시킨다는 것은, 처음에는 데이터를 입력받아 전혀 엉뚱한 값을 내뱉던 모델을, 점점 정답에 해당하는 값을 출력할 수 있도록 '보정'해나가는 과정이라고도 할 수 있습니다.

모델을 학습시키는 방법은 모델 형태에 따라 다양합니다. 딥러닝의 신경망 모델은 'Gradient Descent(경사하강법)'이라는 방법을 사용하죠.

# 14-12. AI의 현재와 미래 (1) 머신러닝과 딥러닝


머신러닝은 주어진 데이터로부터 패턴을 학습해서 원하는 값을 예측하는 것에 방점이 있다면, 딥러닝은 입력된 데이터에 내재되어 있는 표현 그 자체를 나타내도록 학습하는 것이 목표입니다. 다음 이미지로 그 차이가 명확하게 드러나죠.

기존의 머신러닝은 데이터를 입력하기 위해 사람이 직접 피처(Feature)를 가공해 주어야 했습니다. 전통적인 머신러닝 모델들은 주로 '표' 형태로 된 정형 데이터를 처리하도록 설계되었기 때문입니다. 그래서 지금까지 이미 해봤거나 앞으로도 그럴 예정이듯, 머신러닝 모델을 학습시킬 때에는 데이터를 전처리하고 모델에 들어가도록 이쁘게 가공하는 작업이 필수적입니다. 또한, 주어진 데이터를 이리저리 조합해서 더욱 좋은 피처를 만들어내는 것도 성능을 높이기 위한 굉장히 중요한 과정이죠. 이러한 과정을 통틀어 피처 엔지니어링(Feature Engineering) 이라고 합니다. 피처 엔지니어링은 그 자체로 머신러닝 내에서 한 챕터를 차지할 만큼 중요하고 방대한 내용이 많은 분야입니다. 그만큼 머신러닝은 더 좋은 모델을 만들기 위해 '사람의 개입'을 굉장히 많이 필요로 합니다.

한편, 딥러닝은 모델의 복잡성을 늘려 사람이 개입하는 피처 엔지니어링을 최소화하고자 합니다. 이미지라면 픽셀 형태로 된 2차원 배열을 그대로 신경망에 입력하면, 알아서 그 안에 있는 특징이나 표현을 학습하도록 하죠. 물론 자연어라면 한글이나 알파벳을 넣을 순 없기에 자연어를 숫자로 이루어진 벡터 형태로 가공하는 작업은 해야 하지만(자연어 임베딩(Embedding)) 그럼에도 머신러닝과는 조금 다릅니다. 머신러닝으로는 자연어를 입력받아 기계 번역을 한다던가, 문장을 생성하는 등의 태스크를 수행할 수 없죠. 이처럼 딥러닝은 뚜렷한 형태가 없는(비정형) 데이터로부터 표현을 추출해 내는 것에 매우 강합니다.

딥러닝에서는 이렇게 데이터를 모델에 입력해서 바로 원하는 출력값을 얻어내는 End-to-End Learning을 추구하는 경우가 많습니다. 끝에서 끝까지 모두 담당한다는 정도의 의미가 되겠네요.
